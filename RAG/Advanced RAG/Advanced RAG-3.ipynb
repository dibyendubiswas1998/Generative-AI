```python
# Documents:

documents = [
    "This is a list which containing sample documents.",
    "Keywords are important for keyword-based search.",
    "Document analysis involves extracting keywords.",
    "Keyword-based search relies on sparse embeddings.",
    "Understanding document structure aids in keyword extraction.",
    "Efficient keyword extraction enhances search accuracy.",
    "Semantic similarity improves document retrieval performance.",
    "Machine learning algorithms can optimize keyword extraction methods.",
    "Machine learning (ML) is defined as a discipline of artificial intelligence (AI) that provides machines the ability to automatically learn from data and past experiences to identify patterns and make predictions with minimal human intervention.",
    "There are four types of machine learning algorithms: supervised, semi-supervised, unsupervised and reinforcement.",
    "Statistics is a mathematical body of science that pertains to the collection, analysis, interpretation or explanation, and presentation of data, or as a branch of mathematics. ",
    "Descriptive and Inferential Statistics.",
    "Hi, I am Dibyendu Biswas, learning Hybrid-Search, Keyword Search, Vector Search."
]
```

```python
from sentence_transformers import SentenceTransformer
```

```python
# Load pre-trained Sentence Transformer model for creating embeddings:

model_name = 'sentence-transformers/paraphrase-xlm-r-multilingual-v1'
model = SentenceTransformer(model_name)
```

```python
# Create Embeddigs based on documents:

document_embeddings = model.encode(documents)
```

```python
document_embeddings[0]
```

```python
# Len of embeddings:

len(document_embeddings[0])
```

```python
# Create query embeddings:

query = "Natural language processing techniques enhance keyword extraction efficiency."
query_embedding = model.encode(query)
query_embedding
```

```python
len(query_embedding)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# get the Similarity scores based on query_embeddings:

similarities = cosine_similarity(np.array([query_embedding]), document_embeddings)
similarities
```

```python
# Get the top most similarity index

most_similar_index = np.argmax(similarities)
most_similar_index
```

```python
# Get the corresponding documents:

documents[most_similar_index]
```

```python
# Sorted the indexes based on similarity scores:

sorted_indices = np.argsort(similarities[0])[::-1]


# And then get the Ranked Documents:
ranked_documents = [(documents[i], similarities[0][i]) for i in sorted_indices]
ranked_documents
```

```python
print("Ranked Documents:")
for rank, (document, similarity) in enumerate(ranked_documents, start=1):
    print(f"Rank {rank}: Document - '{document}', Similarity Score - {similarity}")
```

```python
# Print Top 5 Ranked Documents:

print("Ranked Documents:")
for rank, (document, similarity) in enumerate(ranked_documents[:5], start=1):
    print(f"Rank {rank}: Document - '{document}', Similarity Score - {similarity}")
```

## **Perform Re-Ranking:**

### **Using BM25:**

```python
!pip install rank_bm25 -q
```

```python
from rank_bm25 import BM25Okapi
```

```python
# Get the Top 4 Documents:

top_4_documents = [doc[0] for doc in ranked_documents[:4]]
top_4_documents
```

```python
# Tokenized the top_4-documents by splitting:

tokenized_top_4_documents = [doc.split() for doc in top_4_documents]
tokenized_top_4_documents
```

```python
# Tokenized the query also:

query = "Natural language processing techniques enhance keyword extraction efficiency."
tokenized_query = query.split()
tokenized_query
```

```python
# Apply BM25 also on tokenized_top_4_documents

bm25 = BM25Okapi(tokenized_top_4_documents)
bm25
```

```python
# get scores based on tokenized_query:

bm25_scores = bm25.get_scores(tokenized_query)
bm25_scores
```

```python
# Get the index by soring bm25_scores:

sorted_indices2 = np.argsort(bm25_scores)[::-1]
sorted_indices2
```

```python
# Get the Re-Ranked Documents:

reranked_documents_bm25 = [(top_4_documents[i], bm25_scores[i]) for i in sorted_indices2]
reranked_documents_bm25
```

```python
# Print Re-Rankings Documents:

print("Rerank of top 4 Documents:")
for rank, (document, similarity) in enumerate(reranked_documents_bm25, start=1):
    print(f"Rank {rank}: Document - '{document}', Similarity Score - {similarity}")
```

```python
# get the model for embeddings:

cross_encoder = CrossEncoder('cross-encoder/ms-marco-MiniLM-L-6-v2')
```

```python
# Get the Top 4 Ranked Documents:

ranked_documents_ = top_4_documents
ranked_documents_
```

```python
# Query:

query = "Natural language processing techniques enhance keyword extraction efficiency."
```

```python
# Create Pairs of query & documents:

pairs = []
for doc in top_4_documents:
    pairs.append([query, doc])
pairs
```

```python
# Get the scores, by applying Cross Encoder:

scores = cross_encoder.predict(pairs)
scores
```

```python
scored_docs = zip(scores, top_4_documents)
```

```python
# Get the Re-Ranked Documents by applying Cross Encoder:

reranked_document_cross_encoder = sorted(scored_docs, reverse=True)
reranked_document_cross_encoder
```
